# Ordenar lista: programa depurado

```python
list = []

for i in range(0,10):
  num = random.randint(0,100)
  #print(num)

  list.append(num)

print(list)
```

```python
print('list ini=',list)

for k in range(len(list)):
  #Sacar el menor de una lista
  x1= list[k] 
  for i in range(len(list)-k):
    if list[i+k] <= x1:
      x1= list[i+k]
  xmin=x1

  #posición del menor econtrado
  for i in range(len(list)):
    if list[i] == xmin:
      j=i
  #print('j=',j)

  #permutar el minimo al principio
  a=list[k] #guardo list[k]
  list[k]=list[j]
  list[j]=a
  
print('list fin=',list)
```

Resumen: Lo que hace este programa es buscar el número menor de la lista, colocarlo al principio de la lista. Luego buscar el segundo menor y colocarlo en el segundo lugar, buscar el tercer menor y colocarlo en el tercer lugar, etc. Si el menor buscado está en la posición donde debe quedar, entonces no hace nada.
