# **Operadores o estructuras de decisión**

Ya que hemos jugado un poco con Python, es hora de entrar más en detalle con las estructuras de decisión.

Una estructura de decisión hace un tipo de selección (escoger algo). En otras palabras, escoge que instrucciones ejecutar del programa y cuales no.  
Existe la palabra reservada **if**, **elif** y **else** (si, de lo contrario). 

La estructura  if/elif/else es una forma común de controlar el flujo de un programa, lo que te permite ejecutar bloques de código específicos según el valor de algunos datos.

# Sentencia **if**

La estructura de decisión **if** realiza una acción específica sólo cuando la condición es verdadera; de lo contrario, la acción se ignora.

Ejemplo:

```python
numero = 3

if numero == 2:
  print('La variable numero es 2')
  print("entro al if")

print('prosigue el programa')  
```

```python
nota = 1.0

if nota >= 3.0:  print('Aprobó el curso')

print('\n Prosigue el código normalmente...')
```


# Sentencia **if/else** **(SI / DE LO CONTRARIO)**

La estructura **else** permite especificar que una acción diferente se debe desarrollar, cuando la condición es falsa.

Ejemplo:

```python
numero = float(input())
```

```python
if numero == 2.0:
  print('El numero ingresado es 2')

else:
  print("El número ingresado es distinto de 2, el numero fue",numero)  
```

```python
#ejemplo básico.
nota = 2.0

if nota >= 3.0:
  print('Aprobó el curso')

else:
  print('Perdió el curso')  
```

# Sentencia **if/elif/else**

Estructura de selección múltiple **if/elif/else**. Se evalua el **i**f, sino se cumple se evalua el **elif** y si éste último no se cumple se evalue el **else** (se pueden poner varios elif)

Ejemplo:

```python
n = 30

if n == 1:
  print('n es 1')

elif n ==2:
  print('n es 2')

elif n == 3:
  print('n es 3')  

else:
  print('El numero no es ni 1, ni 2 ni 3')  
```

**Ejemplo:** Redondeo de un número a una cifra decimal (teoría estadística)


REGLAS:

1.   Si la segunda cifra decimal es menor que 5 se elimina. Por ejemplo: $2.94 \to 2.9$
2.   Si la segunda cifra decimal es mayor que 5, entonces la cifras inmediatamente anterior se aumenta en una unidad. Por ejemplo: $2.86 \to 2.9$
3. Si la segunda cifra decimal es 5, entonces se evalua la primera cifra decimal: Si es par se deja igual, si es impar, se incrementa en una unidad. Por ejemplo: $2.65 \to 2.6$ y $2.75 \to 2.8$



```python
#### PROGRAMA HECHO EN CLASE 
x = 2.75

#Multiplica por 100 para tomar las tres primeras cifras
y = int(x*100)

#Descompone en unidades, decenas y centenas
uni = y%10
dec = int((y%100 - uni)/10)
cen = y//100

#print(' los numero son x',x,'y',y,'uni=',uni,'dec=',dec,'cen=',cen)

if uni < 5:
  resultado = (y -uni)/100 
  print('resultado=',resultado)

if uni > 5:
  resultado = (cen*100 +(dec+1)*10)/100  
  print('resultado=',resultado)

if ( uni == 5 and dec%2 == 0):
  resultado = (y -uni)/100 
  print('resultado=',resultado)

else:
  resultado = (cen*100 +(dec+1)*10)/100  
  print('resultado=',resultado)

```
