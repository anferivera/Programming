# Ejemplo: Grafiquemos una función

```python
#Preparar
X = np.linspace(0, 30, 50)
Y = X**2
```

```python
#Render
fig, ax = plt.subplots()
ax.plot(X,Y,color='red',linestyle='-.',linewidth=2)
fig.show()
```

Ejemplo: Hagamos dos gráficas completas

```python
### Tamaño de la figura
plt.figure(figsize=(6,4))

X = np.linspace(-2,2,50)
Y =X**4
Z = X**3

#Forma abreviada
plt.plot(X,Y,linestyle='-',marker='.',label='cuad')
plt.plot(X,Z,color='red',marker='*',label='cub')

#Titulo
plt.title('Función polinómica',size='26')
#Ejes
plt.xlabel('x (m)',size='14')
plt.ylabel('y=x$^n$ (m)',size='14')
# grid y legenda
plt.grid()
plt.legend(loc='lower right')
#guardar figura (desccomente la sigueinte línea)
#plt.savefig('polinomio.pdf')
plt.show()
```

```python
#Tamaño de la figura
plt.figure(figsize=(6,4))

X = np.linspace(0, 10*np.pi, 50)
Y = X**2

plt.plot(X,Y, 'r.')
plt.plot(X,Y, color='blue')

#Título
plt.title('Función cuadrática')
#Ejes
plt.ylabel('x$^2$',size='18')
plt.xlabel('x (m)',size='18')
plt.grid()

#plt.savefig('cuadratica.pdf')
plt.show()
```

### Función Meshgrid

La función numpy.meshgrid devuelve una lista de matrices de coordenadas a partir de vectores de coordenadas.

```python
coord_x = [0,1,2]
coord_y = [0,1,2]
x, y = np.meshgrid(coord_x,coord_y)
```

```python
print(x)
print(y)
```

Como vemos, la función acepta dos vectores de coordenadas, coord_x , coor_y. Devuelve dos arrays, 'x' e 'y'. El primero conteniene la coordenada 'x' de los 9 puntos, y el segundo la coordenada 'y' de los mismos. Una vez obtenidos estos arrays, es posible evaluar una función sobre los puntos.

```python
z = np.sqrt(x**2+y**2)
print(z)
```

Ejemplo real:

```python
coord = np.arange(-5,5,0.01) #100 puntos
x , y = np.meshgrid(coord, coord, indexing= 'ij')
z = np.sqrt(x**2+y**2)*np.sin(x**2+y**2)
```

Hemos considerado como vector de coordenadas 1000 puntos entre los valores -5 y 5, y a las matrices de coordenadas resultantes les hemos aplicado cierta función cuyo resultado se ha almacenado en la variable z, tanto 'x' como 'y' son arrays NumPy de tamaño (1000, 1000), 'z' también es un array NumPy del mismo tamaño.

```python
x.shape
```

```python
print(type(z))
print(z.shape)
```

```python
plt.figure(figsize= (8,6))
plt.imshow(z, cmap = "summer")
plt.colorbar()
plt.show()
```

## His:

Calcula y dibuja un histograma.

El método usa numpy.histogram para partir los datos en x y contar el numero de vlores en cada partición (bin). Luego dibuja la distribución como un (BarContainer or Polygon).
Los bines , rango, density y peso son dados a numpy.histogram.

*`matplotlib.pyplot.hist(x, bins=None, range=None, density=False, weights=None, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, color=None, label=None, stacked=False, *, data=None, **kwargs) `*


https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html


```python
np.random.normal(0,1)
```

```python
#distrubución normal (gaussiana) con denviación estándar $\sigma=0.1$
#https://en.wikipedia.org/wiki/Normal_distribution
Z = np.random.normal(0,0.1,1000)

plt.hist(Z,bins=30)

plt.grid()

plt.show()
```

https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html

```python
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)

count, bins, ignored = plt.hist(s, 50, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.grid()
plt.show()
```

## pie:

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html
