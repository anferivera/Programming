# **Operadores lógicos**


\begin{array}{|c|c|c|}
\hline
\color{red}{\text{Python}} &  & \\
\hline
\text{not} & ! & \text{Negación}\\
\text{and} & \&\& & \text{Conjunción}\\
\text{or} & || & \text{Disyunción}\\
\hline
\end{array}

En las variables lógicas. 0 representa falso y 1 representa verdadero.

Prioridad de los operadores lógicos.


1.   Negación
2.   Conjunción
3.   Disyunción








Ejemplos:

```python
2 == 3
```

```python
 not(2 == 3)
```

```python
(2<3) and (5<4)
```

```python
(2<3) and (2<1)
```

```python
(7<3) or (10<1)
```

```python
2 == 3 or 2<3
```

Ejemplo: Entrera tres numeros (a,b,c) por teclado y calcular a==b or a==c 

```python
print('Entre la variable numérica a=')
a = input(float())
print('Entre la variable numérica b=')
b = input(float())
print('Entre la variable numérica c=')
c = input(float())

print('\n***********************************')
print('Las variables entradas fueron:')
print('a=',a,'b=',b,'c=',c)
print('*************************************\n')

(a == b) or (a == c)
```
