## Sintaxis

En Python, una función tiene las siguientes características:

1.   La palabra clave def
2.   Un nombre de función
3.   Paréntesis ’()’,
4.   Dentro de los paréntesis los parámetros de entrada, aunque los parámetros de entrada sean opcionales.
5.   Dos puntos ’:’
6.   Algún bloque de código para ejecutar
7.   Una sentencia de retorno 'return' (opcional)

EJEMPLO: Código de la burbuja hecho en la clase 09. Ahora el código lo juntamos todo en una función llamada bubble_sort que recibe una lista como argumento.

```python
def bubble_sort(list):

  for i in range(len(list)-1):
    swapped = False #Variable para saber si no hay nigún intercambio y parar porque hay orden

    for j in range(len(list)-1):

      if list[j] > list[j+1]:
        a = list[j+1] #guardo lo que hay a la der
        list[j+1] = list[j] # rescribo a la derecha
        list[j] = a #rescribo a la izquierda

        swapped = True

    if swapped == False:
      break

  return list
```

```python
x = [64, 81, 13, 65, 91, 97, 84, 16, 28, 1]

bubble_sort(x)
```

EJEMPLO: Función sen(x)

\begin{align}
\sin(x) = \sum_{n=0}^{∞}\dfrac{(-1)^n}{(2n+1)!}x^{2n+1} = x - \dfrac{x^3}{1.2.3}+\dfrac{x^5}{1.2.3.4.5}- \dfrac{x^7}{1.2.3.4.5.6.7} \cdots
\end{align}

$|x|<1$

```python
def seno(x):

  seno= x - x**3/(1*2*3) + x**5/(1*2*3*4*5)- x**7/(1*2*3*4*5*6*7)+ x**9/(1*2*3*4*5*6*7*8*9)

  return seno
```

```python
seno(0.1)
```
