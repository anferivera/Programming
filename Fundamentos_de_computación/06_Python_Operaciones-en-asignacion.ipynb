Ejemplos básicos:

```python
num=7 
print('num=',num) 

num = num+3;  
print("num=", num) 
```

```python
num=7 
print('num=',num) 

num += 3;  
print("num=", num) 
```

```python
x=7 
print('x=',x) # 7

x+=2;  
print("x+=", x) # 9
```

Ejemplo: escribir los múltiplos de 2

```python
i=0; j=0;

for i in range(0,10):

  j += 2
  print('i=',i,'->',j)
```

```python
a=7; b=2
print("Operadores de asignación")
x=a; x+=b;  print("x+=", x)  # 9
x=a; x-=b;  print("x-=", x)  # 5
x=a; x*=b;  print("x*=", x)  # 14
x=a; x/=b;  print("x/=", x)  # 3.5
x=a; x%=b;  print("x%=", x)  # 1
x=a; x//=b; print("x//=", x) # 3
x=a; x**=b; print("x**=", x) # 49
x=a; x&=b;  print("x&=", x)  # 2
x=a; x|=b;  print("x|=", x)  # 7
x=a; x^=b; print("x^=", x)   # 5
x=a; x>>=b; print("x>>=", x) # 1
x=a; x<<=b; print("x<<=", x) # 28
```
