```python
#Crear la lista
list = []

for i in range(0,10):
  num = random.randint(0,100)

  list.append(num)

print(list)
```

Coloca el xmin al principio

```python
print(list)
#Sacar el menor de una lista
xmin=min(list)

#posición del menor econtrado
for i in range(len(list)):
  if list[i] == xmin:
    j=i
print('j=',j)

#permutar el minimo al principio
a=list[0] #guardo x0
list[0]=list[j]
list[j]=a
print(list)
```
