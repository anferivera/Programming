## Ejemplos: FISICA

1.   Ejemplo 1: Caida libre


$y = h +v_0t-\dfrac{1}{2}gt^2$ 

$v_y = v_0 -g t$

si $v_0=0$ el tiempo de caída se encuentra cuando y=0. $t=\sqrt{2h/g}$ 

EJEMPLO: Un cuerpo se deja caer desde un edificio de altura h. Escribamos un programa que calcule la posición (y) y la velocidad de caida cada 0.1 s hasta que impacte el piso. Escribir una tabla con tres columnas (t,y,v).

````python
from IPython.display import Image
img = 'caidalibre.png'
Image(filename=img)
````

````python
import numpy as np

f = open("caidalibre.txt", "w")

h = 100.
g = 9.8
tc = (2*h/g)**(1/2)
print(tc)
tiempo = np.arange(0,tc,0.1)

for t in tiempo:

  y = h -(1./2.)*g*(t)**2 
  v = g*t #rapidez

  #print('t(s)=', "%.2f" % t,'   y(m)=', "%.2f" % y,'   v(m/s)=',"%.2f" % v)

  f.write("%.2f " % t + "%.2f " % y +"%.2f\n" % v )

f.close()
````

````python
f = open("caidalibre.txt", "r")
print(f.read())
f.close()
````

WARNING: A manera de información hacemos la siguiente gráfica. Luego estudiaremos en detalle Numpy, matplotlib, tec. Sólo es por ilustración

````python
import pandas as pd
import matplotlib.pyplot as plt

names1 = ['tiempo', 'y','v']
a = np.loadtxt('caidalibre.txt')
cl = pd.DataFrame(a, columns=names1)
````

````python
cl[:3]
````

````python
plt.figure(figsize=(6,4))

plt.plot(cl.tiempo, cl.y, 'r.', label='caida libre')

plt.legend(prop={'size':8}, loc='upper right')
plt.grid()

plt.tight_layout()
plt.show()
````

````python
plt.figure(figsize=(6,4))

plt.plot((cl.y*0 +1), cl.y, 'r.', label='caida libre')

plt.legend(prop={'size':8}, loc='best')

plt.xlim(0.98,1.02)
plt.grid()
plt.ylabel('Altura')

plt.tight_layout()
plt.show()
````
