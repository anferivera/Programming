## Arrays

NumPy trabaja con una estructura de datos llamada array o arreglos numéricos multidimensionales. 
Un array unidimensional puede ser una fila o una columna de una tabla. erstos arrays son llamados **ndarray** 

1.   Un array unidimensional puedes ser una fila o columna de una tabla, este se conoce como **vector** (rango 1). 
2. Un array bidimensional es lo que llamamos **matriz** (rango 2).
3. Y un array de 3 dimensiones (o más), es decir, se conoce como **tensor**.

### **Manipulando arrays:** índices, slicing,...

Los ndarrays permiten hacer slicing (rebanadas) con la notación [filas, columnas] y sus índices inician en 0.

Para seleccionar toda una columna o fila, se usa **:**, tal que:


*   [n, :] ← esto selecciona todas las columnas de la n-esima fila.
*   [:, n] ← selecciona todas las filas de la n-ésima columna.

Vemaos algunos ejemplos:



```python
a = np.random.randint(0,10,size=(6,6))

a
```

```python
x = []
for i in range(6):
  
  y = a[0,i]
  print(y)
  x.append(y)

x = np.array(x)
```

```python
x
```

```python
#Imprimir el elemento[1,2] (fila 1, segunda y columna 2 tercera). Python cuenta desde 0.
a[1,2]
```

```python
# Imprimir todos los datos de la fila 2
a[0,:]
```

```python
a[2]
```

```python
a
```

```python
# Imprimir todos los datos de la columnda 3
a[:,3]
```
