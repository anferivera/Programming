# **Precedencia de operadores**



Python aplica los operadores según un orden determinado por su prioridad. A continuación se muestran los operadores de más prioritarios a menos prioritarios.

see: https://interactivechaos.com/es/manual/tutorial-de-python/precedencia-de-operadores

![](https://drive.google.com/uc?export=view&id=1eigWp9dOraSeTN56XJyILnU7NB_LzfNP)

Ejemplo 1:

```python
3*2+2*4
```

```python
3*(2+2)*4
```

# Casilla de texto

Ejemplo 2:

```python
print(2**2*3,2**(2*3),'haciendo dos operaciones')
```

Note la diferencia en el orden al operar. Por ejemplo:

*   Primero hace la multiplicación y luego hace la suma.
*   Primero hace la exponenciación y luego la multiplicación.
